## Phase Advance Filter

Función de transferencia:


$$H(s)=\frac{1}{1+\frac{R_1}{R_2(1+sR_1C)}}$$


Magnitud:


$$|H(j\omega)|=\frac{1}{\sqrt{1+\frac{R_1²+2R_1R_2}{R_2²(1+\omega²R_1²C²)}}}$$


Argumento:


$$arg(H(j\omega))=tan^{-1}(\omega R_1C)-tan^{-1}(\frac{\omega R_1R_2C}{R_1+R_2})$$

```julia
MAG(ω)=1/sqrt(1+((r1^2+2*r1*r2)/(r2^2*(1+ω^2r1^2c^2))))
```

```julia
ARG(ω)=atan(ω*r1*c)-atan((r1*r2*ω*c)/(r1+r2))
```

```julia
r1=1.5e3
r2=300
c=150e-12
```

```julia
plot(x->MAG(x),0,2/(r1*c),label="",c="red",title="Bode Plot log-log |H(s)|")
xaxis!(:log10)
yaxis!(:log10)
```

```julia
plot(x->ARG(x)-π/2,0,1e9,label="",c="red",title="Bode Plot linear-log arg(H(iw))")
xaxis!(:log10)
```

## PI Filter

```julia
MAGF(ω)=1/(sqrt(1+(2r1*r2+r1^2)/(r2^2+(1/(ω^2*c^2)))))
```

```julia
ARGF(ω)=atan(r2*c*ω)-atan((r1+r2)*c*ω)
```

```julia
r1=3e3
r2=150
c=47e-9
```

```julia
plot(x->MAGF(x),0,100,label="",c="red",title="Bode Plot log-log |H(s)|")
xaxis!(:log10)
yaxis!(:log10)
```

```julia
plot(x->ARGF(x),0,100,label="",c="red",title="Bode Plot linear-log Arg(H(s))")
xaxis!(:log10)
```
